# **Ridership on Amtrak Trains**

Amtrak, a US railway company, routinely collects data on ridership. Here, we focus on forecasting future ridership using the series of monthly ridership between January 1991 and March 2004. Hence, our task here is (numerical) time series forecasting. (Shmueli, 20191028, p. 65)

Shmueli, G., Bruce, P. C., Gedeck, P., Patel, N. R.  (20191028). Data Mining for Business Analytics.

# Data

## Get Data

```python
file_name = 'datasets/Amtrak.csv'

amtrack_df = pd.read_csv(file_name, index_col='Month')
amtrack_df.shape
```

## Clean Data

```python
# Fix data type
amtrack_df.index = pd.to_datetime(amtrack_df.index, format='%d/%m/%Y')
```

```python
amtrack_df.dtypes
```
